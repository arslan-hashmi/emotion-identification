# Binary Class Plot 

```python
import matplotlib.pyplot as plt
import numpy as np

import pandas as pd
data = pd.read_csv('Single\\SVM_Results\\2_Class\\Top_200\\0_accuracy.csv')
data['Label'] = data['Label'].str.replace('1','Happy')
data['Label'] = data['Label'].str.replace('2','Sad')
data['Label'] = data['Label'].str.replace('3','Anger')
data['Label'] = data['Label'].str.replace('4','Fear')
data['Label'] = data['Label'].str.replace('5','Disgust')
data['Label'] = data['Label'].str.replace('6','Surprise')
data.sort_values('Acc')
Average = round(data['Acc'].mean(),2)

Y = data.Acc.astype(float)
N = data.Acc.shape[0]
data2 = pd.read_csv('Single\\RF_Results\\2_Class\\Top_200\\0_accuracy.csv')
data2['Label'] = data2['Label'].str.replace('1','Happy')
data2['Label'] = data2['Label'].str.replace('2','Sad')
data2['Label'] = data2['Label'].str.replace('3','Anger')
data2['Label'] = data2['Label'].str.replace('4','Fear')
data2['Label'] = data2['Label'].str.replace('5','Disgust')
data2['Label'] = data2['Label'].str.replace('6','Surprise')

Y2 = data2.Acc.astype(float)
Average2 = round(data2['Acc'].mean(),2)

X = np.full(N,0)
c = np.random.rand(N), 
L = data.Label.astype(str)
A = np.full(N, 100)


i =0

fig = plt.figure(figsize=(5,5), dpi=300)#
title = "SVM: "+str(Average)+"%"+"   RF: "+str(Average2)+"%"
plt.title(title, size=11)
plt.ylim(82,93)
plt.xlim(-1, 5) 
plt.xticks([0 , 1], ["SVM","RF"])
mark = ['o', '.', 'X', 'x', '+', 'v', '^', '<', '>', 's', 'd', '1', '2', '3', '4']
color = ['b', 'r', 'g', 'c', 'm', 'y', 'k', 'b', 'r', 'g', 'c', 'm', 'y', 'k', 'r']
for marker in L:
    plt.plot(X[i], Y[i], mark[i], label=L[i], color=color[i])
    plt.plot(X[i]+1, Y2[i], mark[i], color=color[i])
    i = i+1
plt.legend(numpoints=1)
plt.ylabel('Accuracy (%)', size=11) 
plt.savefig('CMs\\comp_step_binary.png',bbox_inches='tight', transparent="True", pad_inches=0.05)   
plt.show()
```
